```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

df =pd.read_csv('vgsales.csv')

sns.countplot(x = 'Platform',linewidth=2, edgecolor='black',data = df)
plt.xticks(rotation=80)
plt.yticks(range(0, 2500, 500))
plt.title('Nombre de jeux par console')
plt.show()

```

```python
df_year = df.sort_values(by="Year", ascending = False)
df_year
```

```python
# Trier les données par plateforme et ventes globales
df_world = df.sort_values(by=['Global_Sales'], ascending=False)
df_eu = df.sort_values(by=['EU_Sales'], ascending=False)
df_na = df.sort_values(by=['NA_Sales'], ascending=False)
df_jp = df.sort_values(by=['JP_Sales'], ascending=False)

top10_df = df_world[['Name','Platform', 'Global_Sales']].head(10)
top10_eu = df_eu[['Name','Platform', 'EU_Sales']].head(10)
top10_na = df_na[['Name','Platform', 'NA_Sales']].head(10)
top10_jp = df_jp[['Name','Platform', 'JP_Sales']].head(10)
```

```python
#top10 MONDE
sns.barplot(x="Name", y = "Global_Sales", hue='Platform',data = top10_df)
plt.xticks(rotation=80)
plt.xlabel('Jeux Vidéos')
plt.ylabel('Ventes totales en millions eur')
plt.title('Top 10 des ventes dans le monde')

plt.show()

#top10 EU
sns.barplot(x="Name", y = "EU_Sales", hue='Platform',data = top10_eu)
plt.xticks(rotation=80)
plt.xlabel('Jeux Vidéos')
plt.ylabel('Ventes totales en millions eur')
plt.title('Top 10 des ventes en Europe')
plt.show()

#top10 NA
sns.barplot(x="Name", y = "NA_Sales", hue='Platform',data = top10_na)
plt.xticks(rotation=80)
plt.xlabel('Jeux Vidéos')
plt.ylabel('Ventes totales en millions eur')
plt.title('Top 10 des ventes en Amérique')

plt.show()

#top10 JP
sns.barplot(x="Name", y = "JP_Sales", hue='Platform',data = top10_jp)
plt.xticks(rotation=80)
plt.xlabel('Jeux Vidéos')
plt.ylabel('Ventes totales en millions eur')
plt.title('Top 10 des ventes au Japon')

plt.show()
```

```python
sns.relplot(x='Year', y='Global_Sales', hue ='Name', kind = 'scatter', ci = None, data = (df[df['Global_Sales']>20]));
```

```python
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(12, 8))
sns.barplot(x="Name", y="Global_Sales", hue='Platform', data=top10_df, ax=axs[0, 0])
sns.barplot(x="Name", y="EU_Sales", hue='Platform', data=top10_eu, ax=axs[0, 1])
sns.barplot(x="Name", y="NA_Sales", hue='Platform', data=top10_na, ax=axs[1, 0])
sns.barplot(x="Name", y="JP_Sales", hue='Platform', data=top10_jp, ax=axs[1, 1])

# Configuration des axes et titres
axs[0, 0].set_xlabel('Jeux Vidéos')
axs[0, 0].set_ylabel('Ventes totales en millions eur')
axs[0, 0].set_title('Top 10 des ventes dans le monde')
axs[0, 1].set_xlabel('Jeux Vidéos')
axs[0, 1].set_ylabel('Ventes totales en millions eur')
axs[0, 1].set_title('Top 10 des ventes en Europe')
axs[1, 0].set_xlabel('Jeux Vidéos')
axs[1, 0].set_ylabel('Ventes totales en millions eur')
axs[1, 0].set_title('Top 10 des ventes en Amérique')
axs[1, 1].set_xlabel('Jeux Vidéos')
axs[1, 1].set_ylabel('Ventes totales en millions eur')
axs[1, 1].set_title('Top 10 des ventes au Japon')

# Ajustement des espacements entre les graphiques
plt.tight_layout()

plt.subplots_adjust(top=1, bottom=0.2, hspace=2, wspace=0.3)

for ax in axs.flat:
    ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
    
# Affichage des graphiques
plt.show()
```

```python
#TOP des jeux les plus vendus dans le monde en fonction des platform

from bokeh.io import show
from bokeh.models import HoverTool
from bokeh.plotting import figure
from bokeh.palettes import Spectral10
from bokeh.transform import factor_cmap
from bokeh.models import ColumnDataSource

# Créer un objet ColumnDataSource pour les données
source = ColumnDataSource(top10_df)

# Créer la figure
p = figure(x_range=top10_df['Name'], plot_height=500, plot_width=900, 
           title="Top 10 des jeux vidéo les plus vendus dans le monde", toolbar_location=None, tools="")

# Ajouter les barres
p.vbar(x='Name', top='Global_Sales', width=0.9, source=source, 
       line_color='white', fill_color=factor_cmap('Platform', palette=Spectral10, factors=top10_df['Platform'].unique()))

# Ajouter le hoover
hover = HoverTool(tooltips=[("Jeu", "@Name"), ("Ventes globales", "@Global_Sales"), ("Platform","@Platform")])
p.add_tools(hover)

# Modifier l'apparence
p.xgrid.grid_line_color = None
p.y_range.start = 0
p.xaxis.major_label_orientation = 1.2
p.title.text_font_size = '18pt'
p.xaxis.axis_label = "Jeu"
p.yaxis.axis_label = "Ventes globales"

# Afficher le graphique
show(p)
```

```python

# Aggrégation des données par console et année
year_platform = df.groupby(['Year', 'Platform'])['Global_Sales'].sum().reset_index()

pivot_data = year_platform.pivot('Year', 'Platform', 'Global_Sales')

# Création du heatmap avec seaborn
sns.heatmap(pivot_data, cmap='Reds')
plt.title("Ventes totales depuis 1980")

# Affichage du plot
plt.show()
```

```python
data_grouped = df.groupby(['Year', 'Platform'])['Global_Sales'].sum().reset_index()


# Création du line plot avec seaborn
sns.lineplot(x='Year', y='Global_Sales', hue='Platform', data=(data_grouped[data_grouped['Global_Sales']>50]))
plt.xticks(range(1980, 2025, 10))

# Affichage du plot
plt.show()
```

```python
from bokeh.io import output_notebook, show
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.plotting import figure
from bokeh.transform import factor_cmap
from bokeh.palettes import Category10_10

# Création des sources de données pour Bokeh
source = ColumnDataSource(data_grouped[data_grouped['Global_Sales']>50])

# Création de la figure
p = figure(plot_width=800, plot_height=400, x_axis_label='Year', y_axis_label='Global Sales')

# Ajout des cercles pour chaque plateforme
p.circle(x='Year', y='Global_Sales', size=8, source=source, legend_field='Platform', 
         fill_color=factor_cmap('Platform', palette=Category10_10, factors=source.data['Platform'].unique()))

# Ajout de la légende
p.legend.location = 'top_left'

# Ajout du hover pour afficher le nom des plateformes
hover = HoverTool(tooltips=[('Platform', '@Platform'),('Global_Sales','@Global_Sales')])
p.add_tools(hover)

# Affichage de la figure
show(p)
```
